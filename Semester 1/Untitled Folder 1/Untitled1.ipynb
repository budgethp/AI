```python
import random

def has_duplicate_birthday():
    # Generate a list of random birthdays for 22 people (1 to 365)
    birthdays = [random.randint(1, 365) for _ in range(22)]
    # Check if there's any duplicate birthday
    return len(birthdays) != len(set(birthdays))

def birthday_probability(trials=10000):
    # Run the simulation for the specified number of trials
    duplicate_count = 0
    for _ in range(trials):
        if has_duplicate_birthday():
            duplicate_count += 1
    # Calculate the probability
    return duplicate_count / trials

# Run the simulation and print the probability
print("Theoretical Probability of Shared Birthday in a Group of 22:", birthday_probability())
```

```python
import random

def has_brown_eyed_girl():
    # Simulate whether the child is a girl and has brown eyes
    is_girl = random.random() < 0.5  # 50% chance
    has_brown_eyes = random.random() < 0.75  # 75% chance
    return is_girl and has_brown_eyes

def brown_eyed_girl_probability(trials=100000):
    count_brown_eyed_girls = 0
    for _ in range(trials):
        if has_brown_eyed_girl():
            count_brown_eyed_girls += 1
    # Calculate probability
    return count_brown_eyed_girls / trials

# Run the simulation and print the empirical probability
print("Empirical Probability of Having a Brown-Eyed Girl:", brown_eyed_girl_probability())
```
