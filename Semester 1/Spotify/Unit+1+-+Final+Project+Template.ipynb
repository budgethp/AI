```python
# Write all the code needed for your analysis here. 
# You can add more cells to this section if you would like.
import pandas as pd
import seaborn as sns
import numpy as np
df = pd.read_csv("pokemon.csv", encoding='iso-8859-1')
```

```python
def genCompare(Comparison):
    df_gen1 = df[df["Generation"]== 1]
    df_gen2 = df[df["Generation"]== 2]
    df_gen3 = df[df["Generation"]== 3]
    df_gen4 = df[df["Generation"]== 4]
    df_gen5 = df[df["Generation"]== 5]
    df_gen6 = df[df["Generation"]== 6]
    sns.histplot(data=df_gen3, x = Comparison, bins=50, color = "yellow")
    sns.histplot(data=df_gen1, x = Comparison, bins=50, color = "red")
    sns.histplot(data=df_gen5, x = Comparison, bins=50, color = "blue")
    sns.histplot(data=df_gen2, x = Comparison, bins=50, color = "orange")
    sns.histplot(data=df_gen4, x = Comparison, bins=50, color = "green")
    sns.histplot(data=df_gen6, x = Comparison, bins=50, color = "purple")
```

```python
def standard_deviation(dataframe, column):
    std = dataframe[column].std()
    return std
def avg(col):
    average = col.mean()
    return average
```
