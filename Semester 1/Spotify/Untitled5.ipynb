```python
import pandas as pd
import seaborn as sns
df = pd.read_csv("spotify2023.csv", encoding = 'ISO-8859-1')
df['streams'] = pd.to_numeric(df['streams'], errors = 'coerce')
df_winterspring = df[df["released_month"] <= 6]
df_summerfall = df[df["released_month"] > 6]
print(df_winterspring['streams'].mean())
print(df_summerfall['streams'].mean())
sns.histplot(data = df_winterspring["streams"], bins = 50, color ="orange")
sns.histplot(data = df_summerfall["streams"], bins = 50, color ="blue")
```
