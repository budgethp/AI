```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('estonia.csv')
isdead = {
    'Deaths': ['Alive', 'Dead'],
    'Count': [0, 0],
}
for idx in range(len(df)):
    if df["Survived"][idx] == 1:
        isdead['Count'][0] += 1
    elif df["Survived"][idx] == 0:
        isdead['Count'][1] += 1
ded = pd.DataFrame(isdead)
print(isdead['Count'][0])
print(isdead['Count'][1])
sns.barplot(x="Deaths", y="Count", data=ded)
```
